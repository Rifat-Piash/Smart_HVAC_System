The aim is to reduce energy waste when rooms are unoccupied, which aligns with the company’s digitalization and sustainability goals. By leveraging data from sensors, the model is designed to predict occupancy and help optimize energy usage, ultimately leading to cost savings.</p>

<h3><font color="blue">Dataset Overview:</font></h3>
<p style="font-size:16px;line-height:2.5em;">
The data used in this study comprised three separate datasets:
</p>

<ul style="font-size:16px;line-height:2.5em;">
    <li>datatest with 2,665 observations,</li>
    <li>datatest2 with 9,752 observations, and</li>
    <li>datatraining with 8,143 observations.</li>
</ul>

<p style="font-size:16px;line-height:2.5em;">
Despite the naming convention suggesting a separation between training and test sets, the datasets were combined into a single set, resulting in a total of 20,560 observations across 7 columns. The data exhibited class imbalance, with 76.9% of the observations representing Class 0 (no occupant) and 23.1% representing Class 1 (occupant present). To address this, K-Fold Cross-Validation was employed, ensuring the model’s robustness and balanced performance across different subsets of data.
</p>

<h3><font color="blue">Feature Engineering:</font></h3>
<p style="font-size:16px;line-height:2.5em;">Several feature engineering steps were performed to improve model performance. Log transformations were applied to the Light and CO2 variables due to high skewness. Time-related features, including hour and day_of_week, were extracted from the date column to visualize temporal patterns and also hot-encoded to check if they have influence towards Occupancy. However, after conducting two feature selection methods such as Mutual Information and Extra Trees Classifier, time-related feature was not included in the final feature selection to give the model more generalization power. Variables such as HumidityRatio, Light, CO2, LogCO2 and date were dropped from the dataset after transformation, as they were either redundant or strongly correlated, contributing little to the predictive power of the model after feature engineering. Based on the consistent findings from various analysis, the final selected features for the model were Temperature, Humidity, Log_Light, and rolling_mean_CO2</p>

<h3><font color="blue">Modeling and Evaluation:</font></h3>
<p style="font-size:16px;line-height:2.5em;">After performing feature engineering, including the introduction of a rolling mean for CO2, the data was split into training, testing, and validation sets. Systematic sampling was used to select representative samples for final validation, and K-Fold cross-validation was implemented to ensure robust model evaluation. Initially, five machine learning models—Logistic Regression, Random Forest, Decision Tree, Support Vector Machine (SVM), and XGBoost—were trained and evaluated on both K-Fold cross-validation and unseen validation data. The addition of the rolling mean CO2 feature significantly boosted the models' performance, particularly on unseen data, with accuracy improving from 89% to 100%. This feature helped capture temporal patterns in CO2 levels linked to occupancy, enhancing the model's predictive accuracy.</p>

<p style="font-size:16px;line-height:2.5em;">Hyperparameter tuning was subsequently performed for each model, resulting in further improvements across performance metrics. After tuning, the Random Forest model achieved an average cross-validation accuracy of 99% and an ROC AUC of 0.99, consistently outperforming the other models in stability and accuracy across both cross-validation and unseen data. The Random Forest model also demonstrated strong generalization due to its ability to leverage the newly engineered features effectively, making it the top choice for deployment.</p>

<h3><font color="blue">Recommendation:</font></h3>
<p style="font-size:16px;line-height:2.5em;">The Random Forest model is recommended for deployment, as it delivered the highest and most stable performance across all evaluation metrics. By incorporating the rolling mean CO2 feature, this model captures occupancy patterns more accurately, facilitating real-time detection and optimizing energy usage by controlling HVAC systems based on occupancy. Regular monitoring and scheduled retraining are advised to ensure the model remains accurate over time, adapting to any shifts in building usage patterns.</p>

<h2><a class="anchor" id="Section_2_1"></a><font color=blue>Literature Review</font></h2>
<br/><br/>
<p style="font-size:16px;line-height:2.5em;">
The drive for energy efficiency in commercial buildings has intensified as businesses seek to align with sustainability goals and reduce operational costs. According to a report from the U.S. Department of Energy, buildings account for approximately 41% of the total energy consumption in the United States (Dong et al., 2019). Within this, lighting and HVAC systems alone consume around 70% of the energy in commercial office buildings, contributing significantly to carbon dioxide, sulfur dioxide, and nitrogen dioxide emissions, which exacerbate global warming (Kwok & Lee, 2011). Effective control of environmental variables like temperature, humidity, and lighting in buildings is therefore critical for both energy efficiency and environmental impact reduction.
</p>
<p style="font-size:16px;line-height:2.5em;">
Occupant behavior significantly impacts energy usage in commercial buildings. Typically, HVAC and lighting systems operate on fixed schedules, often disregarding the actual occupancy patterns, which leads to substantial energy wastage. Studies estimate that around 70% of energy used for HVAC and lighting is consumed during periods when rooms are unoccupied (Dong et al., 2019). Fine-grained occupancy data, gathered through advanced sensor technologies, can enable more responsive building control systems, reducing unnecessary energy use by adjusting HVAC and lighting based on real-time occupancy data (Kwok & Lee, 2011).
</p>
<p style="font-size:16px;line-height:2.5em;">
Indoor environmental quality (IEQ) is closely tied to occupant health and productivity. Poorly managed temperature and lighting can lead to discomfort and reduced productivity. For instance, excessively cold environments can cause discomfort and health issues, leading to reduced cognitive performance (Asif & Zeeshan, 2020). Lighting, particularly the quality and intensity of light, also plays a crucial role. Daylight, as recommended by the ASHRAE 90.1 standard, which suggests an illuminance of 500 lux, has been shown to enhance occupant productivity compared to artificial lighting (Dong et al., 2019). Thus, maintaining optimal conditions for IEQ not only benefits energy efficiency but also promotes a healthier and more productive environment for building occupants.
</p>
<p style="font-size:16px;line-height:2.5em;">
Recent advancements in sensor technologies and machine learning models have facilitated intelligent control of building systems, allowing for real-time adjustments in HVAC operations based on occupancy patterns. Studies have demonstrated the potential of occupancy sensors in predicting room occupancy and aligning energy usage accordingly, reducing energy waste significantly (Zhao et al., 2015). Virtual sensors and multi-source information fusion approaches, such as Bayesian belief networks, have been proposed to improve the accuracy of occupancy detection, further enhancing the ability to implement demand-driven HVAC operations (Zhao et al., 2015).
</p>

```python
def findMissingValue(df):
    # Calculate the number of missing values per column
    missing_values = df.isnull().sum()

    # Calculate the percentage of missing values
    percent_missing = (missing_values / len(df)) * 100

    # Create a DataFrame with the results
    missing_data = pd.DataFrame({
        'Columns': df.columns,
        'Number_Of_MValue': missing_values,
        'Percentage_Of_MValue': percent_missing.round(2)
    })

    # Filter out columns with no missing data and sort
    missing_data = missing_data[missing_data['Number_Of_MValue'] > 0].sort_values('Number_Of_MValue', ascending=False).reset_index(drop=True)

    return missing_data
```

<h2><a class="anchor" id="Section_5_1"></a><font color=blue>5.1 Train, Test, and Validation Data Split</font></h2>
<br/><br/>
<p style="font-size:16px;line-height:2.5em;">
In this approach, a systematic sampling method is used to create a validation dataset that captures a range of conditions within the data, ensuring that the model is validated on representative and critical scenarios. Specifically, the function data_index_for_validation identifies samples with extreme values (maximum and minimum) for key features — Temperature, Humidity, Log_Light, and rolling_mean_CO2 — under both occupancy conditions (Occupied and Not Occupied). This selection method ensures that the validation set includes diverse edge cases, allowing for a robust evaluation of the model's performance under varying conditions. After creating the validation set, the remaining data is used as the training set, ensuring that the model is trained on a comprehensive dataset while being tested on meaningful and challenging examples. This systematic validation split supports reliable model assessment by including significant feature variations.
</p>
<br/><br/>
<a href="#tbc">Go Back To Table Of Content</a>


```python
# Creating a function that will extract some samples from the dataset to validate the model

def data_index_for_validation(df):
    
    cols = ['Temperature', 'Humidity', 'Log_Light', 'rolling_mean_CO2']
    idx = []

    for i in cols:
        # Maximum value with Occupancy == 1
        max_occ_1 = df[(df[i] == df[i].max()) & (df['Occupancy'] == 1)].index
        if not max_occ_1.empty:
            idx.append(max_occ_1[0])

        # Maximum value with Occupancy == 0
        max_occ_0 = df[(df[i] == df[i].max()) & (df['Occupancy'] == 0)].index
        if not max_occ_0.empty:
            idx.append(max_occ_0[0])

        # Minimum value with Occupancy == 1
        min_occ_1 = df[(df[i] == df[i].min()) & (df['Occupancy'] == 1)].index
        if not min_occ_1.empty:
            idx.append(min_occ_1[0])

        # Minimum value with Occupancy == 0
        min_occ_0 = df[(df[i] == df[i].min()) & (df['Occupancy'] == 0)].index
        if not min_occ_0.empty:
            idx.append(min_occ_0[0])

    return idx
```

<h2><a class="anchor" id="Section_5_2"></a><font color=blue>5.2 Building Machine Learning Models</font></h2>
<br/><br/>
<p style="font-size:16px;line-height:2.5em;">
Building machine learning models such as Random Forest, Decision Tree, Logistic Regression, Support Vector Machine, and XGBoost. The performance of each model will be evaluated to track progress and assess improvements achieved before and after hyperparameter tuning.

```python
# Building and Evaluating Models with K-Fold Cross-Validation and Testing on Unseen Data

# List of models to evaluate
models = [
    ("Random Forest", RandomForestClassifier()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Support Vector Machine", SVC(probability=True)),
    ("XGBoost", XGBClassifier())
]

# Data structure to store all results
results = []
unseen_data_results = []

# Building model function
def building_model(models):
    for model_name, model in models:
        print(f"\nEvaluating Model: {model_name}")
        
        # Parameters
        k = 5  # Number of folds

        # Initialize Stratified K-Fold
        skf = StratifiedKFold(n_splits=k)

        # Lists to store performance metrics for the current model
        accuracy_list = []
        roc_auc_list = []
        precision_list = []
        recall_list = []
        f1_list = []

        # K-Fold Cross-Validation
        for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
            print(f"\nFold {fold + 1}/{k}")

            # Split data into training and testing sets
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            # BalancedBaggingClassifier with the current model as base estimator
            bbc_model = BalancedBaggingClassifier(
                base_estimator=model,
                sampling_strategy='auto',
                replacement=False,
                random_state=1
            )

            # Train the classifier
            bbc_model.fit(X_train, y_train)

            # Make predictions
            predictions = bbc_model.predict(X_test)

            # Evaluate performance metrics
            accuracy = bbc_model.score(X_test, y_test)
            roc_auc = roc_auc_score(y_test, predictions)
            report = classification_report(y_test, predictions, output_dict=True)

            precision = report['weighted avg']['precision']
            recall = report['weighted avg']['recall']
            f1 = report['weighted avg']['f1-score']

            # Append metrics for each fold
            accuracy_list.append(accuracy)
            roc_auc_list.append(roc_auc)
            precision_list.append(precision)
            recall_list.append(recall)
            f1_list.append(f1)

            # Print the metrics for each fold
            print(f"Accuracy: {accuracy:.4f}")
            print(f"ROC AUC: {roc_auc:.4f}")
            print(f"Precision: {precision:.4f}")
            print(f"Recall: {recall:.4f}")
            print(f"F1-Score: {f1:.4f}")

        # Calculate the average metrics for the current model
        avg_accuracy = np.mean(accuracy_list).round(4)
        avg_roc_auc = np.mean(roc_auc_list).round(4)
        avg_precision = np.mean(precision_list).round(4)
        avg_recall = np.mean(recall_list).round(4)
        avg_f1 = np.mean(f1_list).round(4)

        # Append results to the results list
        results.append({
            'Model': model_name,
            'Average Accuracy': avg_accuracy,
            'Average ROC AUC': avg_roc_auc,
            'Average Precision': avg_precision,
            'Average Recall': avg_recall,
            'Average F1-Score': avg_f1
        })

        # Test on unseen data
        idx = data_index_for_validation(df)  # Custom function to get indices for unseen data
        test = df.iloc[idx].copy()
        validate = test.drop(columns='Occupancy')

        # Save and load the model using pickle
        saved_model = pickle.dumps(bbc_model)
        model_from_pickle = pickle.loads(saved_model)

        # Make predictions on unseen data
        validate['Prediction'] = model_from_pickle.predict(validate)
        validate['Occupancy'] = test['Occupancy']
        val_result = validate[['Prediction', 'Occupancy']].copy()
        val_result.reset_index(drop=True, inplace=True)
        val_result['Total'] = val_result['Prediction'] + val_result['Occupancy']
        
        # Calculate accuracy on unseen data
        unseen_accuracy = (len(val_result[(val_result['Total'] == 2) | (val_result['Total'] == 0)]) / len(val_result))

        # Append unseen data accuracy to unseen_data_results list
        unseen_data_results.append({
            'Model': model_name,
            'Unseen Data Accuracy': unseen_accuracy
        })

        # Print the average metrics for the model
        print("\n=== K-Fold Cross-Validation Results ===")
        print(f"Model: {model_name}")
        print(f"Average Accuracy: {avg_accuracy}")
        print(f"Average ROC AUC: {avg_roc_auc}")
        print(f"Average Precision: {avg_precision}")
        print(f"Average Recall: {avg_recall}")
        print(f"Average F1-Score: {avg_f1}")
        print(f"Unseen Data Accuracy: {unseen_accuracy:.2f}%")

# Run the function with the list of models
building_model(models)

# Convert results to a DataFrame for summary
results_df = pd.DataFrame(results)
unseen_results_df = pd.DataFrame(unseen_data_results)
```

```python
# # Building XGBoost Model
# Parameters
k = 5  # Number of folds

# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=k)

# Lists to store performance metrics
accuracy_list = []
roc_auc_list = []
precision_list = []
recall_list = []
f1_list = []

# K-Fold Cross-Validation
for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    
    print(f"Fold {fold + 1}/{k}")
    
    # Split data into training and testing sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]   
 
    # Step 3: BalancedBaggingClassifier with XGBoost as base estimator
    bbc_xg = BalancedBaggingClassifier(
        base_estimator=XGBClassifier(kernel= 'linear',gamma=.1,C=.1),
        sampling_strategy='auto',
        replacement=False,
        random_state=1
    )
    
    # Train the classifier
    bbc_xg.fit(X_train, y_train)
    
    # Make predictions
    predictions = bbc_xg.predict(X_test)
    
    # Evaluate performance metrics
    accuracy = bbc_xg.score(X_test, y_test)
    roc_auc = roc_auc_score(y_test, predictions)
    report = classification_report(y_test, predictions, output_dict=True)
    
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1 = report['weighted avg']['f1-score']
    
    # Append metrics to list
    accuracy_list.append(accuracy)
    roc_auc_list.append(roc_auc)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)
    
    # Print the metrics for each fold
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("\n")

model_name.append('XGBoost')
model_accuracy.append(np.mean(accuracy_list).round(4))
model_roc_auc.append(np.mean(roc_auc_list).round(4))
model_precision.append(np.mean(precision_list).round(4))
model_recall.append(np.mean(recall_list).round(4))
model_f1_score.append(np.mean(f1_list).round(4))  
    
# After cross-validation, print average performance
print("=== K-Fold Cross-Validation Results ===")
print(f"Average Accuracy: {np.mean(accuracy_list):.4f}")
print(f"Average ROC AUC: {np.mean(roc_auc_list):.4f}")
print(f"Average Precision: {np.mean(precision_list):.4f}")
print(f"Average Recall: {np.mean(recall_list):.4f}")
print(f"Average F1-Score: {np.mean(f1_list):.4f}")
```
